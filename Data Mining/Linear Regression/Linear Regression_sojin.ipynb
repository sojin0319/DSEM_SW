TV: advertising dollars spent on TV for a single product in a given market (in thousands of dollars) \
Radio: advertising dollars spent on Radio \
Newspaper: advertising dollars spent on Newspaper \
Sales: sales of a single product in a given market (in thousands of widgets)

```python
theta0 = np.random.rand(1)   # 랜덤으로 0과 1사이의 난수를 theta0, theta1 저장
theta1 = np.random.rand(1) 
print('파라미터 값 세타0: ',theta0 ,'세타1: ' , theta1)
```

```python
X = data['TV']
Y = data['Sales']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X,Y,test_size = 0.3, random_state = 0)
```

```python
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
Y_train = Y_train.reset_index(drop=True)
Y_test = Y_test.reset_index(drop=True)
```

```python
train_num = 10000  # 학습 횟수
loss=[] # 손실함수 리스트
a=0.00001  # 학습률
for i in range(train_num):  # 학습 횟수 만큼 반복
    total_loss = 0
    for j in range(len(X_train)):  # 데이터 개수 만큼 반복
        theta0 += a*(Y_train[j] - (theta0+theta1*X_train[j])) * 1          # theta0 갱신
        theta1 += a*(Y_train[j] - (theta0+theta1*X_train[j])) * X_train[j]  # theta1 갱신
        total_loss += ((theta1*X_train[j]+theta0-Y_train[j])**2) / len(X_train)        # 손실 함수
    print('반복횟수: ',i+1,'세타0: ',theta0 ,'세타1: ' , theta1)
```

```python
print('최종 오차값: ',total_loss) # MSE 평균제곱오차 값
```

```python
print('추정된 파라미터의 값 세타0: ',theta0 ,'세타1: ' , theta1) 
```

```python
X = data.drop(['Sales'],axis=1)
Y = data['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split (X,Y,test_size = 0.3, random_state = 0)
```

```python
X_train = X_train.reset_index(drop=True)
X_test = X_test.reset_index(drop=True)
Y_train = Y_train.reset_index(drop=True)
Y_test = Y_test.reset_index(drop=True)
```

```python
w = np.random.rand(3,1)
b = np.random.rand(1) 
print('파라미터 값 w: ',w ,'b: ' , b)
```

```python
def gradient(x):
    g =0
    for i in range(len(w)):
        g += w[i]*x[i]
    return g
```

```python
train_num = 15000  # 학습 횟수
loss=[] # 손실함수 리스트
a=0.00001  # 학습률
for i in range(train_num):  # 학습 횟수 만큼 반복
    total_loss = 0
    for j in range(len(X_train)):  # 데이터 개수 만큼 반복
        b += a*(Y_train[j] - (b+gradient(X_train.loc[j]))) * 1          # theta0 갱신
        m = a*(Y_train[j] - (b+gradient(X_train.loc[j]))) * X_train.loc[j]
        w += m.values.reshape(-1,1)
        total_loss += ((w*X_train.loc[j].values.reshape(-1,1)+b-Y_train.loc[j])**2) / len(X_train)        # 손실 함수
    print('반복횟수: ',i+1,'세타0: ',b ,'세타1: ' , w, 'loss:',np.sum(total_loss)/len(X_train))
    loss.append(total_loss)   # 손실함수 리스트에 추가
```

```python
print('최종 오차값: ',np.sum(total_loss) / len(X_train)) # MSE 평균제곱오차 값
```

```python
def predict(x):
    y = np.dot(x,w)+b
    return y
```
