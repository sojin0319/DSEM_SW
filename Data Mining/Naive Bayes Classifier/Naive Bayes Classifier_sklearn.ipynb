```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
import numpy as np
import seaborn as sns
data= pd.read_csv(r"./dataset/Iris.csv") 
data
```

```python
a = data[data['Species'] == 'Iris-versicolor'].index
data = data.drop(a)
data
```

```python
species_mapping = {"Iris-setosa": 0, "Iris-virginica": 1, "Iris-versicolor":2}
data['Species'] = data['Species'].map(species_mapping)
data
```

```python
X = data.iloc[:, [3, 4]].values
Y = data['Species'].values
Y
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,Y_train)
predict = model.predict(X_test)
```

```python
iris_test = pd.DataFrame({
                        'PetalLength':[1.0],
                        'PetalWidth':[0.1]})
predict = model.predict(iris_test)
predict
```

```python
plot_decision_regions(X=X,y=Y, clf=model, legend=4)
```
